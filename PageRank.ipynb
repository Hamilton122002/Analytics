```python
import networkx as nx
import pylab as plt
d = nx.DiGraph()
d.add_weighted_edges_from([('A','B',1), ('A','C',1), ('C','A',1), ('B','C',1), ('B','A',1), ('A','D',1), ('B','D',1), ('D','A',1), ('C','D',1), ('D','B',1)])
print (nx.pagerank(d))
```

```python
nx.draw(d, with_labels=True)
plt.show()
```
